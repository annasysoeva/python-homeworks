# Задание 1. Базовое изучение

Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
# выбираю необходимые стобцы
horse_table = horse_info[['surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'peripheral pulse', 'pain', 'outcome']]
horse_table
```

# Задание 2. Работа с выбросами

Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
# нахожу выбросы для столбца возраста через межквартильный размах
q1 = horse_table.Age.quantile(0.25)
q3 = horse_table.Age.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_table[horse_table.Age.between(lower_bound, upper_bound, inclusive=True)]
pd.concat([horse_table, remove_outliers]).drop_duplicates(keep=False)

# Скорее всего, это ошибка ввода данных, т.к. для возраста может быть только два значения: 1 = Adult horse 2 = Young (< 6 months)
# Если в задаче этот стобец будет необходим, то мы не сможем работать с такими данными
# Тогда придется их заменить или не учитывать при рассчете, если остальной информации будет достаточно
```

```python
# температура
q1 = horse_table['rectal temperature'].quantile(0.25)
q3 = horse_table['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_table[horse_table['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([horse_table, remove_outliers]).drop_duplicates(keep=False)

# Видим два показателя высокой температуры и один низкой. 
# Скорее всего, это естественные выбросы, поэтому трогать их не будем
```

```python
# пульс
q1 = horse_table.pulse.quantile(0.25)
q3 = horse_table.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_table[horse_table.pulse.between(lower_bound, upper_bound, inclusive=True)]
pd.concat([horse_table, remove_outliers]).drop_duplicates(keep=False)

# Судя по документации, нормальный показатель пульса - до 40 ударов
# На результаты выше этого показателя нужно смотреть по ситуации.
# Это могут быть естественные вбросы из-за болезни,но может быть и ошибка.
```

```python
# частота дыхания
q1 = horse_table['respiratory rate'].quantile(0.25)
q3 = horse_table['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_table[horse_table['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([horse_table, remove_outliers]).drop_duplicates(keep=False)

# Нормальные показания - от 8 до 10
# Скорее всего, показатели 84 и 70 это ошибки
```

```python
# периферический пульс
q1 = horse_table['peripheral pulse'].quantile(0.25)
q3 = horse_table['peripheral pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse_table[horse_table['peripheral pulse'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([horse_table, remove_outliers]).drop_duplicates(keep=False)

# Здесь только пропуски
```

```python
for col in horse_table.columns:
    pct_missing = horse_table[col].isna().mean()
    print(f'{col} - {pct_missing :.1%}')
    
# пропуски есть во всех стобцах, кроме столбца с возврастом
```

```python
# в столбце с операциями я решила поставить замену на 0
# т.к. мы не можем точно предсказать была операция или нет, а ложное значение может испортить результат при решении вопросов
# та же ситуация с исходом - мы не знаем, умерла лошадка или нет

filled_table = horse_table.fillna({'surgery':0,'outcome':0})
```
